```python
import random
import time

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]

def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]

def merge_sort(arr):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_arr = arr[:mid]
        right_arr = arr[mid:]
        merge_sort(left_arr)
        merge_sort(right_arr)

        i = j = k = 0
        while i < len(left_arr) and j < len(right_arr):
            if left_arr[i] < right_arr[j]:
                arr[k] = left_arr[i]
                i += 1
            else:
                arr[k] = right_arr[j]
                j += 1
            k += 1

        while i < len(left_arr):
            arr[k] = left_arr[i]
            i += 1
            k += 1

        while j < len(right_arr):
            arr[k] = right_arr[j]
            j += 1
            k += 1

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr)//2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def test_sorting_algorithm(sort_func):
    # Ask the user to input the number of elements in the array
    n = int(input("Enter the number of elements in the array: "))

    # Generate a random array of integers
    arr = [random.randint(1, 1000) for _ in range(n)]

    # Measure the time taken to sort the array
    start_time = time.time()
    sort_func(arr)
    end_time = time.time()
    elapsed_time = end_time - start_time

    # Determine the space complexity of the sorting algorithm
    space_complexity = len(arr)

    return elapsed_time, space_complexity

# Test bubble sort
bubble_sort_time, bubble_sort_space = test_sorting_algorithm(bubble_sort)
print("Bubble sort time: {} seconds".format(bubble_sort_time))
print("Bubble sort space complexity: {} units".format(bubble_sort_space))

# Test selection sort
selection_sort_time, selection_sort_space = test_sorting_algorithm(selection_sort)
print("Selection sort time: {} seconds".format(selection_sort_time))
print("Selection sort space complexity: {} units".format(selection_sort_space))

# Test merge sort
merge_sort_time, merge_sort_space = test_sorting_algorithm(merge_sort)
print("Merge sort time: {} seconds".format(merge_sort_time))

# Test quick sort
quick_sort_time, quick_sort_space = test_sorting_algorithm(quick_sort)
print("Quick sort time: {} seconds".format(quick_sort_time))
print("Quick sort space complexity: {} units".format(quick_sort_space))
# Test insertion sort
insertion_sort_time, insertion_sort_space = test_sorting_algorithm(insertion_sort)
print("Insertion sort time: {} seconds".format(insertion_sort_time))
print("Insertion sort space complexity: {} units".format(insertion_sort_space))

```
